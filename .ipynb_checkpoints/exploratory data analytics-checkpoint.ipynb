```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
import math
from geopy.geocoders import Nominatim
```

```python
#load the data
df=pd.read_csv("lady_footlocker_locations_feb_2021.csv")
df
```

```python
#info about data
df.info()
```

```python
#Making a full address of if address is null replace with address 1
df["Full Address"]=df["address_line_1"]+df["address_line_2"]
df["Full Address"]=df["Full Address"].fillna(df["address_line_1"])
df.isnull().sum()
```

```python
geolocator = Nominatim(user_agent="ak.adarshkaul@gmail.com")
location=[]
lat=[]
long=[]
null_values=[]
for i in range(0,59):
    location.append(geolocator.geocode(df["Full Address"].iloc[i]))
    
```

```python
#if loaction is unavilabe then
df["location"]=location
df["zip_code"]=df["zip_code"].astype(str)
zip=df["zip_code"]
city=df["city"]
city_zip=zip+ ","+city
df["location"]=df["location"].fillna(city_zip)
```

```python
import pgeocode
nomi = pgeocode.Nominatim('US')
for i in range(0,59):
    lat.append(nomi.query_postal_code(df["zip_code"].iloc[i]).latitude)
    long.append(nomi.query_postal_code(df["zip_code"].iloc[i]).longitude)
```

```python

# Python program to plot  
# geographical data using plotly 
  
# importing all necessary libraries 
import plotly.plotly as py 
import plotly.graph_objs as go 
import pandas as pd 
  
# some more libraries to plot graph 
from plotly.offline import download_plotlyjs, init_notebook_mode, iplot, plot 
  
# To establish connection 
init_notebook_mode(connected = True) 
  
  
# type defined is choropleth to 
# plot geographical plots 
data = [ go.Scattergeo(
        locationmode = 'USA-states',
        lon =long,
        lat = lat,
        mode = 'markers',
        marker = dict( 
            size = 10,
            opacity = 0.8
        ))]
layout = dict(
        title = 'Business locations', 
        geo = dict(
            scope='usa',
            projection=dict(type = 'albers usa'),
            showland = True,
            landcolor = "rgb(250, 250, 250)",
            subunitcolor = "rgb(217, 217, 217)",
            countrycolor = "rgb(217, 217, 217)",
            countrywidth = 0.5,
            subunitwidth = 0.5        
        ),
    )
# passing data dictionary as a list  
fig= go.Figure(data = data, layout = layout) 
  
# plotting graph 
iplot(fig)
#image saving
fig.write_image("/home/akaul/Desktop/Projects/exploraity data analysis/exploraity data analysis/fig1.png")
```
